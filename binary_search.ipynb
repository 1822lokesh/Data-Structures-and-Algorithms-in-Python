#  Binary Search
Binary Search is a searching algorithm that works on sorted arrays (ascending or descending).

Instead of checking elements one by one (like Linear Search), it divides the search space in half at each step.
This makes it much faster than Linear Search.



 **How it Works (Step-by-Step)**

Suppose we want to search for 30 in this sorted array:

[10, 20, 30, 40, 50]

1. Start with low = 0 and high = n-1 ‚Üí (0, 4).
2. Find the mid index:
mid = (low + high) // 2

    ‚Üí (0+4)//2 = 2

3. Check arr[mid]:

   If arr[mid] == target, return index.

   If arr[mid] > target, search the left half.

   If arr[mid] < target, search the right half.

Repeat until low > high (means element not found).

**üëâ In our case**:

mid = 2 ‚Üí arr[2] = 30 ‚Üí ‚úÖ Found at index 2.

**‚è± Time Complexity**

 * Best Case: O(1) ‚Üí target is the middle element.

 * Worst Case: O(log n) ‚Üí keep dividing until one element left.

 *  Average Case: O(log n)

**Space Complexity:**

  *  Iterative ‚Üí O(1)
* Recursive ‚Üí O(log n) (stack calls)


```python
list=[10,20,30,40,50]
low=0
high=len(list)-1
key=int(input("enter a key: "))
found=False
while low <=high :
        mid=(low+high)//2
        if key==list[mid]:
            print("key found")
            found=True
            break
        elif key < list[mid]:
             high=mid-1
        else:
             low=mid+1
if not found:
    print("Element not found")









```
